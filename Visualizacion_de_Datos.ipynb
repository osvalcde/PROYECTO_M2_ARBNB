### Analisis y tratado de daotos.

```python
#exportar librerias necesarias
%matplotlib inline
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.io as pio
#import ace_tools as tools
import io
```

### Lectura archivo CSV en un DataFrame y revisión los primeros registros.
* Este paso se realizara con cada archivo .csv

```python

df = pd.read_csv('data/calendar.csv')
# Muestra las primeras filas del DataFrame
print(df.head())

print("\nInformación general del DataFrame:")
print(df.info())

print("\nEstadísticas descriptivas de columnas numéricas:")
print(df.describe())

print("Cantidad de valores nulos por columna:")
print(df.isnull().sum())
```

```python
# Calcular el porcentaje de valores nulos por columna
null_percentage = (df.isnull().sum() / len(df)) * 100

# Crear gráfico de barras para visualizar los porcentajes de valores nulos
plt.figure(figsize=(10, 6))
bars = plt.bar(null_percentage.index, null_percentage, color='skyblue')

# Añadir etiquetas con el porcentaje sobre cada barra
for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        f'{bar.get_height():.1f}%',
        ha='center', va='bottom'
    )

# Etiquetas y título
plt.xlabel('Columnas')
plt.ylabel('Porcentaje de Nulos')
plt.title('Porcentaje de Valores Nulos por Columna')
plt.xticks(rotation=45)
plt.ylim(0, 100)

# Mostrar gráfico
plt.tight_layout()
plt.show()
```
